```python
def double_resol(net, x):
    # x = 1 image - C, H, W
    C, H, W = x.shape
    input_res = 32
    n_H = H // input_res
    n_W = W // input_res
    patches = [torch.chunk(t, n_W, dim=2) for t in torch.chunk(x, n_H, dim=1)]
    patches = [item for sublist in patches for item in sublist]
    patches = torch.stack(patches)
    z = torch.randn((len(patches), 3, 64, 64), dtype=torch.float32)
    x2, _ = net(z, patches, reverse=True)
    x2 = make_grid(x2, n_H, padding=0)
    x2 = torch.sigmoid(x2).detach()
    return x2
```
